```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# X: your original DataFrame
# y: your target variable
# rfe: your fitted RFE object

# 1. Select the 10 features chosen by RFE
X_selected = X.loc[:, rfe.support_]

# 2. Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# 3. Initialize and train XGBoost regressor with GPU support
model = xgb.XGBRegressor(
    objective='reg:squarederror',
    tree_method='hist',  # Use 'hist' and set device='cuda' for GPU as of XGBoost 2.0+
    device='cuda',
    random_state=42
)
model.fit(X_train, y_train)

# 4. Predict and evaluate
y_pred = model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Test RMSE: {rmse:.4f}")
print(f"Test MAE: {mae:.4f}")
print(f"Test R²: {r2:.4f}")
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

# Assume you have: X_train, y_train, X_val, y_val, X_test, y_test

# Combine train and validation for grid search if needed, or use only train for fitting and validation for early stopping

# Define parameter grid
param_grid = {
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'n_estimators': [100, 300],
    'gamma': [0, 1]
}

# Initialize the XGBRegressor with GPU and early stopping
xgb_reg = xgb.XGBRegressor(
    objective='reg:squarederror',
    tree_method='hist',      # Use 'hist' with device='cuda' for GPU
    device='cuda',           # Specify GPU device
    early_stopping_rounds=10, # Set early stopping here, not in .fit()
    random_state=42,
    verbosity=1
)

# GridSearchCV will refit on the whole training set; pass eval_set for early stopping
# Use a lambda to inject eval_set into fit_params
fit_params = {
    "eval_set": [(X_val_scaled, y_val)],
    "verbose": True
}

grid_search = GridSearchCV(
    estimator=xgb_reg,
    param_grid=param_grid,
    scoring='neg_root_mean_squared_error',
    cv=3,
    verbose=2,
    n_jobs=-1
)

# Run grid search with fit_params for early stopping
grid_search.fit(X_train_scaled, y_train, **fit_params)

print("Best parameters:", grid_search.best_params_)
print("Best CV RMSE:", -grid_search.best_score_)

# Evaluate on the test set
y_pred = grid_search.best_estimator_.predict(X_test_scaled)
test_rmse = mean_squared_error(y_test, y_pred, squared=False)
print("Test RMSE:", test_rmse)
```

```python
# After running grid_search.fit(...)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best hyperparameters found:", best_params)

# Get the best estimator (model with those parameters)
best_model = grid_search.best_estimator_

# Optionally, get the best cross-validation score
best_cv_score = -grid_search.best_score_  # Negated because scoring was 'neg_root_mean_squared_error'
print("Best cross-validation RMSE:", best_cv_score)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Predict on the test set
y_pred = best_model.predict(X_test_scaled)

# Calculate RMSE
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

# Calculate MAE
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.2f}")

# Calculate R² Score
r2 = r2_score(y_test, y_pred)
print(f"R² Score: {r2:.2f}")
```
